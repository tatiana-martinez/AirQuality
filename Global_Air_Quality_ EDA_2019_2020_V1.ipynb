### Import CSV data files into Python

We store all csv files in google drive. We use a function to collect data from urls of google drive and store them in one dataframe.

```python
# Function to get the data from google drive

def get_data(org_url):
    file_id = org_url.split('/')[-2]
    dwn_url='https://drive.google.com/uc?export=download&id=' + file_id
    url = requests.get(dwn_url).text
    csv_raw = StringIO(url)
    data = pd.read_csv(csv_raw) #latin1  
    return data
```

```python
# List of urls link to the csv files in google drive

org_url1 = 'https://drive.google.com/file/d/1X8wWfyMJm1VzuCfXMMBzb-UmzJtG1JOC/view?usp=sharing'
org_url2 = 'https://drive.google.com/file/d/1uY04U1DoQM5jl4AI5NmXBI63GPvjp38j/view?usp=sharing'
org_url3 = 'https://drive.google.com/file/d/1V4kpWQ9cHthntyyHFOU_XCMq586TgOX5/view?usp=sharing'
org_url4 = 'https://drive.google.com/file/d/1H5P4a678ig-JWGe7gGZgyqNT9j1nwkKj/view?usp=sharing'
org_url5 = 'https://drive.google.com/file/d/1UuqZuCT9urSOfm0ZribIzFAK-baJVXAL/view?usp=sharing'

lst_url = [org_url1,org_url2,org_url3,org_url4,org_url5]
```

```python
# Use the get_data function to collect the data and store them in a dataframe

df = []

for n in lst_url:
    data = get_data(n)
    df.append(data)
    
df = pd.concat(df)

df.head()
```

#### Add new columns

Based on the country code, we will add two new columns: Country_name and Continent for the sake of our further analysis

```python
# by using the function pycountry, we create a dictionary that contains pairs of code and name of each country

countries = {}
for country in pycountry.countries:
    countries[country.name] = country.alpha_2

# We define a function to get the name of country based on the country code
def get_country_name(x):
        for name, code in countries.items(): 
            if code == x:
                return name

# We define a function to get the name of continent based on the country code
def get_continent(x):
    continent_code = pc.country_alpha2_to_continent_code(x)
    continent_name = pc.convert_continent_code_to_continent_name(continent_code)
    return continent_name

# We apply these functions above to create the 2 new columns called "Country_name" and "Continent"
df['Country_name'] = df['Country'].apply(lambda row: get_country_name(row))
df['Continent'] = df['Country'].apply(lambda row: get_continent(row))
df = df[['Date','Continent','Country','Country_name','City','Specie','count','min','max','median','variance']]
```

#### Analyse missing values

Overview of missing values:

```python
df.isna().sum()/df.shape[0]
```

```python
df.isna().sum()
```

As we can see in the series above, the 'aqi' columns in every index level 0 have the most important % of missing values (97%).

According to the website aqicn.org where we obtained the dataset, we can fill the missing values of AQI by taking the maximum value among those following: PM2.5, PM10, O3 (Ozone), CO (Carbon Monoxide), SO2 (Sulfur Dioxide), NO2 (Nitrogen Dioxide) 

```python
#We save a copy of the original dataframe
df_copy = df.copy()
```

```python
col_lv = ['min','max','median','variance'] #list of column labels at level 0 
```

```python
#We use For loops that allow to replace NaN values and 0 in the columns AQI in differents groups of index. 
for n in col_lv:
    df.loc[(df[n,'aqi'].isna())|(df[n,'aqi'] ==0),[(n,'aqi')]] = df[(df[n,'aqi'].isna())|(df[n,'aqi'] ==0)][(n)][['aqi','no2','so2','pm10','pm25','o3','co']].max(axis=1)
```

```python
df.info()
```

The numbers of non-null values in columns AQI have increased to 312 793 rows now.

As for columns 'wind-speed' and 'wind speed', we note that they contain same type information and complete each other.

Thus, we will group all information in one column and drop the other one.

```python
# Put all information about the wind speed in the column 'wind-speed' and drop the column 'wind speed'
# For loops that allow to replace NaN values in the columns wind-speed by the ones in the columns 'wind speed' 

for n in col_lv:
    df.loc[df[n,'wind-speed'].isna(),[(n,'wind-speed')]] = df[df[n,'wind-speed'].isna()][(n,'wind speed')]
    
df.drop(columns = 'wind speed', level=1, inplace = True)
```

Based on the new dataframe 'df_full', we note that the number of locations is too big that we cannot take into account 100% of locations in our analysis. In addition, the data reported is not completed for a big part of locations (number of days reported < 627 days).


These are reasons why we will reduce the location sample for our analysis by following these steps below:

    * Examine the data contribution by each location by observing the number of rows of each location (country, city)

    * Remove cities that have less than 550 rows. In fact, 550 rows correspond to the number of days from 1.1.2019 to 3.7.2020
    
    * Aggregate cities by country and continent, then select several locations for further analysis.

```python
# Groupby location (country, city) in order to count the rows and we assign this table to a variable

count_row = df_full.groupby(['Continent','Country_name','Country','City']).size().sort_values(ascending=False).reset_index(name='Count')
```

```python
# Extract the list of cities whose values were reported for more than 550 days

count_row[count_row['Count'] > 550]
```

By examining the list above, we note that there are 101 cities whose date was reported for at least 550 days. 

Given that this number of locations is still big, we decide to only take 20 cities from 17 countries of 4 continents that contain a sufficient amount of data for our analysis:

```python
dct_sample = {'Mexico City':'MX','Guadalajara':'MX','Vancouver':'CA','Las Vegas':'US','Jacksonville':'US','São Paulo':'BR','Concepción':'CL','Bogotá':'CO','Sydney':'AU','Brisbane':'AU','Bangkok':'TH','Delhi':'IN','Taipei':'TW','Seoul':'KR','Budapest':'HU','Wrocław':'PL','London':'GB','Hamburg':'DE','Brussels':'BE','Rijeka':'HR'}
```

```python
# We create a function to copy all data relating to the selected locations and paste them into a new dataframe called "df_sample"
df_sample = []

for city,code in dct_sample.items():
    data = df_full[(df_full['Country'] == code) & (df_full['City'] == city)]
    df_sample.append(data)

df_sample = pd.concat(df_sample)
```

#### Analysis of median AQI  before and after the lockdown per city

```python

def graph_median_aqi(city,country):
    #create dataframe the contains the necessary information of the selected city
    df = df_sample[df_sample['City'] ==city]
    
    #get the palette of colors for matplotlib graphs
    palette = plt.get_cmap('Set1')
    
    #Create sub-dataframes for the graph
    x1 = df[df['Date'] < '2020/01/01'][['Date','median_aqi']].sort_values(by = 'Date')
    x2 = df[df['Date'] >= '2020/01/01'][['Date','median_aqi']].sort_values(by = 'Date')
    
    # adjust the size of x and y axis
    plt.figure(figsize = (17,5))
    
    #plot values in the graph
    plt.plot(x1['Date'],x1['median_aqi'], marker='', color=palette(1), linewidth=0.7, alpha=1,label='2019')
    plt.plot(x2['Date'],x2['median_aqi'], marker='', color=palette(6), linewidth=0.7,linestyle='dashed', alpha=1,label='2020')
    plt.legend()
    
    #add label to 2 axes
    plt.xlabel('Period',size = 12)
    plt.ylabel('Median AQI',size = 12)
    
    #add title to the graph
    plt.title(f'{city} - {country}', size = 14)

    return plt.show()
```

We compare the AQI values in the period March to July of both years 2019 and 2020.

The impact of the lockdown is mostly noticeable for:
- In the oceanian continent: 

    * Brisbane : the highest points were from 50 to 75 in the period of 2019, whearas the values AQI fluctuated between 15 to 50 after march 2020.
    * Sydney: range of variation was from 10 to 90 point in 2019 while it was between 10 and 60 points in 2020.

    This decrease was coherent with the lockdown time line of Australia:
    - From March : quarantine applied to travellers comming or returning from abroad (China, Iran, Italy,...)
    - From April:  intrastate travel restriction imposed
